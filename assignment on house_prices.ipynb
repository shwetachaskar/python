```python
import pandas as pd
```

```python
house=pd.read_csv("house_prices.csv")
house.head()
```

```python
house['area_type'].value_counts()
```

```python
x = house.area_type.value_counts().keys().tolist()
print(x)
y = house.area_type.value_counts().tolist()
print(y)
```

```python
import matplotlib.pyplot as plt
plt.bar(x,y)
plt.show()
```

```python
fix_area_type = {"Super built-up  Area":0, "Built-up  Area":1, "Plot  Area":2, "Carpet  Area":3}
house["area_type"]=house.area_type.map(fix_area_type)
```

```python
house.head()
```

```python
def replace_availability(x):
    if x == 'Ready To Move':
        return 0
    elif x == 'Immediate Possession':
        return 1
    else:
        return 2
```

```python
house['availability'] = house.availability.apply(replace_availability)
house.head()
```

```python
house=house.drop(columns="society")
house.head()
```

```python
sizes = house['size'].value_counts().keys().tolist()
size_dictionary = {}
i = 0
for size in sizes:
    size_dictionary[size] = i
    i = i +1
house['size'] = house['size'].map(size_dictionary)
```
