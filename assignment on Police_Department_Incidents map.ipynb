```python
import folium
import pandas as pd
 
SF_COORDINATES = (37.76, -122.45)
crimedata = pd.read_csv('Police_Department_Incidents - Copy.csv')
 
MAX_RECORDS = 50
  

map = folium.Map(location=SF_COORDINATES, zoom_start=12)
 
for each in crimedata[0:MAX_RECORDS].iterrows():
    folium.Marker(location = [each[1]['Y'],each[1]['X']]).add_to(map)
  
display(map)
```
