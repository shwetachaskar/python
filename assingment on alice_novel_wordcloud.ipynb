```python
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
```

```python
text=open("alice_novel.txt").read()
```

```python
text
```

```python
wordcloud=WordCloud(max_font_size=200,max_words=50,background_color="white").generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show
```

```python
STOPWORDS
```

```python
sw=set(STOPWORDS)
```

```python
wordcloud=WordCloud(stopwords=sw,max_font_size=200,max_words=1000,background_color="white").generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show
```

```python
alice_mask = np.array(Image.open("alice_mask.png"))
alice_mask
```

```python
plt.imshow(alice_mask)
```

```python
wordcloud=WordCloud(stopwords=sw,max_words=100,mask=alice_mask).generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show
```

```python
def transform_format(val):
    if val == 0:
        return 255
    else:
        return val
```

```python
transformed_alice_mask = np.ndarray((alice_mask.shape[0],alice_mask.shape[1]), np.int32)

for i in range(len(alice_mask)):
    transformed_alice_mask[i] = list(map(transform_format, alice_mask[i]))
```

```python
transformed_alice_mask
```

```python
wordcloud=WordCloud(stopwords=sw,max_words=500,mask=transformed_alice_mask).generate(text)
plt.figure(figsize=(10,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show
```

```python
wordcloud=WordCloud(stopwords=sw,max_words=500,background_color="white",mask=transformed_alice_mask,contour_width=1,contour_color='black').generate(text)
plt.figure(figsize=(10,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show
```

```python
wordcloud.to_file("alice.png")
```
